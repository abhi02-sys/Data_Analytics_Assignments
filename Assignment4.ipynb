```python
#Random dataset
import pandas as pd
df = pd.read_csv("Random_dataset.csv")
df.head()
```

```python
df = df.mask((df >= 10) & (df <= 60)) 
df
```

```python
#NAs per row
na_per_row = df.isna().sum(axis=1)
print("NAs per row: ", na_per_row)

#NAs per column
na_per_col = df.isna().sum(axis=0)
print("\nNAs per column: ", na_per_col)
```

```python
#Replacing NA with avg of col value
df_new = df.fillna(df.mean())
df_new.head()
```

```python
#heatmap
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(12,8))
sns.heatmap(df_new, cmap="YlGnBu")
plt.title("Heatmap of Dataset")
plt.show()
```

```python
import numpy as np

corr_matrix = df_new.corr().abs()
#daigonals set to 0 to avoid counting self corr
np.fill_diagonal(corr_matrix.values, 0)

count = (corr_matrix <= 0.7).all().sum()
print("Number of columns with all correlations <= 0.7:", count)
```

```python
#Normalize bw 0 n 10
df_norm = df_new.copy()
df_norm = (df_norm - df_norm.min()) / (df_norm.max() - df_norm.min()) * 10
print(df_norm.head())
```

```python
df_binary = df_norm.map(lambda x: 0 if x <= 5 else 1)
print(df_binary.head())
```

```python
plt.figure(figsize=(12,8))
sns.heatmap(df_binary, cmap="YlGnBu")
plt.title("Heatmap of Dataset")
plt.show()
```
