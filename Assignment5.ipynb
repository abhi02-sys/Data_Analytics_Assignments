```python
import numpy as np
#Ratings
customer_A = np.array([4,5,2,3,4])
customer_B = np.array([5,3,2,4,5])
#preferences
customer_A_binary = np.array([1, 0, 1, 1, 0, 1])
customer_B_binary = np.array([1, 1, 1, 0, 0, 1])

#Euclidean dist
euclidean_dist = np.sqrt(np.sum((customer_A - customer_B)**2))
print("Euclidean Distance:", euclidean_dist)

# 2. Manhattan Distance
manhattan_dist = np.sum(np.abs(customer_A - customer_B))
print("Manhattan Distance:", manhattan_dist)

# 3. Cosine Similarity
cosine_sim = np.dot(customer_A, customer_B) / (np.linalg.norm(customer_A) * np.linalg.norm(customer_B))
print("Cosine Similarity:", cosine_sim)

# 4. Hamming Distance
hamming_dist = np.sum(customer_A_binary != customer_B_binary)
print("Hamming Distance:", hamming_dist)

# 5. Jaccard Similarity
intersection = np.sum(np.logical_and(customer_A_binary, customer_B_binary))
union = np.sum(np.logical_or(customer_A_binary, customer_B_binary))
jaccard_sim = intersection / union
print("Jaccard Similarity:", jaccard_sim)
```

```python
#2. Movie Rating
# Movie ratings
user1 = np.array([5, 3, 4, 4, 2])
user2 = np.array([4, 2, 5, 4, 3])

# 1. Chebyshev Distance
chebyshev_dist = np.max(np.abs(user1 - user2))
print("Chebyshev Distance:", chebyshev_dist)

# 2. Minkowski Distance
p = 3
minkowski_dist = np.sum(np.abs(user1 - user2)**p) ** (1/p)
print(f"Minkowski Distance (p={p}):", minkowski_dist)

```
