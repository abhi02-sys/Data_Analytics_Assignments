```python
#To add all nums divisible by 7 and 9
n = int(input("Enter number: "))
total = 0
for i in range(1, n + 1):
    if i % 7 == 0 and i % 9 == 0:
        total += i

print("Sum of numbers divisible by 7 and 9 from 1 to", n, "is:", total)
```

```python
n = int(input("Enter number: "))
total = 0

for num in range(2, n + 1):
    is_prime = True
    for i in range(2, int(num ** 0.5) + 1):
        if num % i == 0:
            is_prime = False
            break
    if is_prime:
        total += num

print("Sum of all prime numbers from 1 to", n, "is:", total)
```

```python
n = int(input("Enter number: "))

def isPrime(n):
    total = 0
    for num in range(2, n + 1):
        is_prime = True
        for i in range(2, int(num ** 0.5) + 1):
            if num % i == 0:
                is_prime = False
                break
        if is_prime:
            total += num
    return total

print("Sum of all prime numbers from 1 to", n, "is:", isPrime(n))
```
