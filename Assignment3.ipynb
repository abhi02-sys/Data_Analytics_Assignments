```python
import pandas as pd
customers = pd.read_csv("AWCustomers.csv")
customers.columns
```

```python
customers.head()
```

```python
sales = pd.read_csv("AWSales.csv")
sales.columns
```

```python
merged_df = pd.merge(customers, sales, on='CustomerID', how='inner')
merged_df.columns
```

```python
selected_cols = ["Education","BirthDate","NumberCarsOwned","Occupation","MaritalStatus","HomeOwnerFlag","TotalChildren","YearlyIncome","BikeBuyer"]
#Filters out selected cols from merged one
selected_df = merged_df[selected_cols]
selected_df.head()
```

```python
#2.
avg_income = selected_df.groupby('BikeBuyer')['YearlyIncome'].mean()
avg_income
```

```python
#percentage bike buyers by occupation
percentage_bikebuyers = selected_df.groupby('Occupation')['BikeBuyer'].mean() * 100
print(percentage_bikebuyers)
```

```python
from datetime import datetime
import pandas as pd

# Ensure BirthDate is in datetime format
selected_df.loc[:, 'BirthDate'] = pd.to_datetime(selected_df['BirthDate'], errors='coerce')

# Compute Age
selected_df.loc[:, 'Age'] = datetime.now().year - selected_df['BirthDate'].dt.year
selected_df.head()
```

```python
avg_age = selected_df.groupby('BikeBuyer')['Age'].mean()
print("Average Age by Bike Buyer status:\n", avg_age)
```

```python
marital_bike = selected_df.groupby('MaritalStatus')['BikeBuyer'].mean() * 100
print("Percentage of Bike Buyers by Marital Status:\n", marital_bike)
```
