```python
#Average salary
avg_salary = df['Salary'].mean()

#Total bonus
total_bonus = df['Bonus'].sum()

# Youngest employee's age
youngest_age = df['Age'].min()

# Highest performance rating
highest_rating = df['Rating'].max()

print("Average Salary:", avg_salary)
print("Total Bonus:", total_bonus)
print("Youngest Age:", youngest_age)
print("Highest Rating:", highest_rating)
```

```python
#sort by salary
df_sorted = df.sort_values(by='Salary', ascending=False)
df_sorted.head()
```

```python
#performance category col added
def categorize_performance(rating):
    if rating >= 4.5:
        return 'Excellent'
    elif rating >= 4.0:
        return 'Good'
    else:
        return 'Average'

df['Performance_Category'] = df['Rating'].apply(categorize_performance)
df.head()
```

```python
#Finding missing valuse
df.isnull().sum()
```

```python
# More than 5 years experience
exp_employees = df[df['Years_of_Experience'] > 5]

#Belong to IT department
it_employees = df[df['Department'] == 'IT']

exp_employees, it_employees
```

```python
#Adding Tax column
df['Tax'] = df['Salary'] * 0.10
df.head()
```

```python
#Saving modified dataset
df.to_csv("modified_employees.csv", index=False)
print("Modified dataset saved successfully")
```
